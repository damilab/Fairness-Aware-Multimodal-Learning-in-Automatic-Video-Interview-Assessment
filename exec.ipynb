```python
import torch

# 16, 3, 224, 224 크기의 랜덤한 텐서 생성
tensor = torch.randn(16, 3, 224, 224)

# 가우시안 노이즈 추가
std_dev = 0.1  # 표준 편차 설정
mean = 0.0     # 평균 설정
noise = torch.randn_like(tensor) * std_dev + mean
noisy_tensor = tensor + noise

# 텐서의 값이 0보다 작으면 0으로 설정
noisy_tensor = torch.clamp(noisy_tensor, min=0.0, max=1.0)

# 출력
print("Noisy Tensor Shape:", noisy_tensor.shape)
```
